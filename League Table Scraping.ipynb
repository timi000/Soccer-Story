```python
LaLiga_list=[]

x=1928

for r in range(8): 
    Lalurl=f"https://www.transfermarkt.com/laliga/tabelle/wettbewerb/ES1?saison_id={x}"
    page=requests.get(Lalurl,headers={'User-Agent': 'Custom'})
    t=pd.read_html(page.content)
    t[3]["Year"]=f"{x+1}-06-30"
    LaLiga_list.append(t[3])
    
    x=x+1
    

#Post Civil War
    
    

x=1939 

for r in range(82):    
    Lalurl=f"https://www.transfermarkt.com/laliga/tabelle/wettbewerb/ES1?saison_id={x}"
    page=requests.get(Lalurl,headers={'User-Agent': 'Custom'})
    t=pd.read_html(page.content)
    t[3]["Year"]=f"{x+1}-06-30"
    LaLiga_list.append(t[3])
    
    x=x+1
    
        
```

```python
LaLiga_list 
```

```python
LaLiga_df=pd.concat(LaLiga_list )
LaLiga_df
LLdf=LaLiga_df.drop(columns='club')
LLdf.columns=['Position','Team','Games_played','Wins', 'Draws',
                                    'Losses','GS_GC','GD','Points','Year']

new = LLdf["GS_GC"].str.split(":", n = 1, expand = True)
LLdf["GS"]= new [0]
LLdf["GC"]= new[1]
pts=LLdf["Points"].str.split(":", n=1, expand=True)
LLdf["PTS"]=pts[0]
LLdf["Position"]=LLdf["Position"].astype(int)
LLdf["Team"]=LLdf["Team"].astype(str)
LLdf["Games_played"]=LLdf["Games_played"].astype(str).astype(int)
LLdf["Wins"]=LLdf["Wins"].astype(str).astype(int)
LLdf["Draws"]=LLdf["Draws"].astype(str).astype(int)
LLdf["Losses"]=LLdf["Losses"].astype(str).astype(int)
LLdf["GD"]=LLdf["GD"].astype(str).astype(int)
LLdf["GS"]=LLdf["GS"].astype(int)
LLdf["GC"]=LLdf["GC"].astype(int)
LLdf["PTS(3PT)"]=(LLdf["Wins"]*3)+LLdf["Draws"]
LLdf["PTS(38G)"]=round((LLdf["PTS(3PT)"]/LLdf["Games_played"])*38,1)
LLdf=LLdf.replace(
    {"Atl. Aviación": "Atlético Madrid","Athletic Madrid": "Atlético Madrid", "Dep. La Coruña":"Deportivo La Coruña", 
  "Ath. Aviación": "Atlético Madrid"  })
LLdf.tail(50)
```

```python
Lteams =pd.DataFrame(LLdf["Team"].value_counts())
Lteams=Lteams.reset_index()
Lteams=Lteams.reset_index()
Lteams["Team_id"]=100+Lteams["level_0"]+1
Lteams["League_id"]=100
Lteams.columns=["old_in", "Teams","Value_count", "Team_id", "League_id"]
Lteams=Lteams[["League_id", "Team_id", "Teams"]]
Lteams
```
