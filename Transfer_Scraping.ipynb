```python
def transf_time(): 
    todays_date=date.today()
    Yeart=todays_date.year
    time_str =( f"{Yeart}-08-15")
    DateT=datetime.strptime(time_str, '%Y-%m-%d').date()
    if todays_date > DateT:
      Yeart=Yeart+1 
    else:
     Yeart=Yeart
    z= Yeart-1995
    return z
```
